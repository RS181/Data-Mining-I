## Working with Indexes

```python
# fb["2019-06"].Close.mean()  fb["2019-06"] da erro
fb.loc["2019-06"].Close.mean()
```

```python
fb["2019-07-05":"2019-07-10"]
```

```python
t=pd.to_datetime("7/22/2019")
t
```

```python
# Escolhe linhas que tenham Date >= 7-22-2019 juntamente com coluna 
fb.loc[fb.index>=t,:]
```
