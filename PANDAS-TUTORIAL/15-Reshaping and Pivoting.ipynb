```python
import pandas as pd
import numpy as np
```

```python
data=pd.DataFrame(
    np.arange(16).reshape(4,4),
    index=[list("aabb"),[1,2]*2],
    columns=[["num","num",
              "comp","comp"],
             ["math","stat"]*2])
data
```

```python
data.index.names=["class","exam"]
data.columns.names=["field","lesson"]
data
```

```python
# data.stack() esta deprecated 
long=data.stack(level=1,future_stack=True)
```

```python
long
```

```python
# unstack move o n√≠vel de √≠ndice especificado (o r√≥tulo da linha) para o eixo da coluna
long.unstack()
```

```python
# stack() esta deprecated ()
# data.stack()
```

```python
# Se o objetivo era empilhar o primeiro n√≠vel do √≠ndice de colunas (podemos usar o abaixo)
data.stack(level=0,future_stack=True)

# Se o objetivo era empilhar o √∫ltimo n√≠vel do √≠ndice de colunas (comportamento padr√£o).
# data.stack(future_stack=True) 

# Se o data for uma Series e o objetivo for desempilhar um n√≠vel.abs
# Use data.unstack(level=0)
```

```python
# esta chamada de stack esta depracated data.stack("field")
data.stack(level="field", future_stack=True)
```

```python
s1=pd.Series(
    np.arange(4),index=list("abcd"))
s2=pd.Series(
    np.arange(6,9),index=list("cde"))
```

## Pivoting ‚ÄúLong‚Äù to ‚ÄúWide‚Äù Format

```python
stock=pd.DataFrame(
    {"fruit": ["apple", "plum","grape"]*2,
     "color": ["purple","yellow"]*3,
     "piece":[3,4,5,6,1,2]})
```

```python
stock
```

```python
# stock.pivot("fruit", "color", "piece")
stock.pivot(index="fruit", columns="color", values="piece")
```

```python
stock["value"]=np.random.randn(len(stock))
```

```python
stock
```

```python
#p=stock.pivot("fruit","color")
p=stock.pivot(index="fruit",columns="color")
p
```

```python
p["value"]
```

## Pivoting ‚ÄúWide‚Äù to ‚ÄúLong‚Äù Format

```python
data=pd.DataFrame(
    {"lesson":["math","stat","bio"],
     "Sam":[50,60,70],
     "Kim":[80,70,90],
     "Tom":[60,70,85]})
data
```

```python
#melt desempilha um conjunto de colunas (que representam diferentes vari√°veis)
# e transforma-as em apenas duas colunas: uma para o nome da vari√°vel (variable)
# e outra para o valor dessa vari√°vel (value).
group=pd.melt(data,["lesson"])
```

```python
group
```

```python
# data=group.pivot("lesson","variable","value") esta forma esta deprecated 
data=group.pivot(index="lesson",columns="variable",values="value")
data
```

```python
data.reset_index()
```
