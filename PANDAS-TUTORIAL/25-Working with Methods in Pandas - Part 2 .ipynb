## to_datetime method

```python
pd.to_datetime("15/08/2019",dayfirst=True)
```

```python
date=["2019-01-05","jan 6, 2019",
      "7/05/2019","2019/01/9","20190110"]
```

```python
# format='mixed' para lidar com formatos inconsistentes
pd.to_datetime(date,format='mixed')
```

```python
pd.to_datetime("03/05/2019")
```

```python
pd.to_datetime("05/03/2019", dayfirst=True)
```

```python
pd.to_datetime("05*03*2019", 
               format="%d*%m*%Y" )
```

```python
date=["2019-01-05",
     "jan 6, 2019",
     "7/05/2019",
     "2019/01/9",
     "20190110"]
```

```python
# format='mixed' para lidar com formatos inconsistentes
pd.to_datetime(date,format="mixed")
```

```python
# pd.to_datetime(date, errors='coerce')
# Usa 'mixed' para inferir o formato de cada elemento e 'dayfirst=True' para resolver a ambiguidade
pd.to_datetime(date, format='mixed', dayfirst=True) 
```

## Frequency and Date Offsets

```python
pd.date_range(
    "2010-01-01","2010-01-03", freq="4h")
```

```python
# freq="WOM-4SUN"
# WOM: Significa "Week Of Month" (Semana do M√™s).
#4: Indica a quarta ocorr√™ncia daquele dia da semana no m√™s.
# SUN: Indica o Dia da Semana (Domingo, em ingl√™s).
pd.date_range(
    "2010-01-01","2010-09-03", 
    freq="WOM-4SUN")
```

```python
rng=pd.period_range(
    "2019-01-01","2019-08-30",freq="M")
rng
```

```python
pd.Series(range(8),index=rng)
```

```python
# Cria um objeto Period que representa o ano inteiro de 2019 (de 1 de janeiro a 31 de dezembro). 
# A √¢ncora do per√≠odo √© Y-DEC (fim do ano em dezembro).
p=pd.Period("2019",freq="Y-DEC")
p
```

```python
# .asfreq('M', ...): Converte a frequ√™ncia do per√≠odo de anual para mensal ('M').
# how='start' alinha o novo per√≠odo mensal com o in√≠cio do per√≠odo original.
p.asfreq('M', how='start')
```

```python
p.asfreq("M",how="end")
```

```python
p=pd.Period("2019Q4", freq="Q-DEC")
p
```

```python
p=pd.Period("2019Q4", freq="Q-FEB")
p
```

```python
p.end_time
```

```python
p.asfreq('D', 'start')
```

```python
rng=pd.period_range(
    "2019Q3","2020Q4", freq="Q-JAN")
rng
```

```python
ts=pd.Series(range(len(rng)),index=rng)
ts
```

```python
rng=pd.date_range(
    "2020-01-01", periods=5, freq="ME")
rng
```

```python
ts=pd.Series(range(len(rng)),index=rng)
ts
```

```python
pts=ts.to_period()
pts.index
```
